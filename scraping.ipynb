# Web Scraping

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys # keys in the keyboard
from selenium.webdriver.common.by import By # used to find elements
```

### Initialize the Chrome driver

```python
driver = webdriver.Chrome('usr/local/bin/chromedriver')
```

### Navigate to URL page

```python
driver.get('https://www.set.or.th/th/market/index/set50/overview')
```

```python
print(driver.title)
```

### Get all html page source

```python
page_source_data = driver.page_source
```

### Convert page source into dataframe

```python
stock_dataFrame = pd.read_html(page_source_data)[2]
```

```python
stock_dataFrame
```

### Clean

```python
stock_dataFrame.columns
```

```python
cleaned_columns = []
for column in stock_dataFrame.columns:
    cleaned_columns.append(column.replace('  (Click to sort Ascending)', ''))
```

```python
cleaned_columns
```

#### Clean column name

```python
stock_dataFrame.columns = cleaned_columns
stock_dataFrame.head()
```

### Set index of the dataframe

```python
stock_dataFrame = stock_dataFrame.set_index('หลักทรัพย์')
stock_dataFrame.head()
```

```python
print(stock_dataFrame.shape)
all_stock_dict = dict()
for stock in stock_dataFrame.index:
    stock_name = stock.split()[0]
    url = f'https://www.set.or.th/th/market/product/stock/quote/{stock_name}/financial-statement/company-highlights'
    driver.get(url)
    stock_data_source = driver.page_source
    table_a = pd.read_html(stock_data_source)[0]
    table_b = pd.read_html(stock_data_source)[1]
    all_stock_dict[stock] = [table_a, table_b]
```
